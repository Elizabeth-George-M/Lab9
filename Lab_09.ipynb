```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import os
import cv2

path = '/content/archive/Training'
classes = {'no_tumor': 0, 'pituitary_tumor': 1}
X = []
Y = []

for cls in classes:
    pth = os.path.join(path, cls)
    for j in os.listdir(pth):
        img = cv2.imread(os.path.join(pth, j), 0)
        img = cv2.resize(img, (200, 200))
        X.append(img)
        Y.append(classes[cls])
```

```python
X = np.array(X)
Y = np.array(Y)

```

```python
pd.Series(Y).value_counts()
```

```python
X_updated = X.reshape(len(X), -1)
```

```python
X.shape, X_updated.shape
```

```python
plt.imshow(X[0], cmap='gray')
```

```python
X_updated = X.reshape(len(X), -1)
X_updated.shape
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=10,
                                               test_size=.20)
```

```python
xtrain.shape, xtest.shape
```

```python
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
xtrain = xtrain/255
xtest = xtest/255
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC()
sv.fit(xtrain, ytrain)
```

```python
print("Training Score:", sv.score(xtrain, ytrain))
print("Testing Score:", sv.score(xtest, ytest))
```

```python
pred = sv.predict(xtest)
```

```python
misclassified=np.where(ytest!=pred)
misclassified
```

```python
print("Total Misclassified Samples: ",len(misclassified[0]))
print(pred[81],ytest[81])
```

```python
dec = {0:'No Tumor', 1:'Positive Tumor'}
```

```python
plt.figure(figsize=(12,8))
c=1
path='/content/archive/Testing/no_tumor'
for i in os.listdir(path)[:9]:
  img=cv2.imread(os.path.join(path,i),0)
  if img is not None:
    p=sv.predict(cv2.resize(img,(200,200)).reshape(1,-1)/255)
    plt.subplot(3,3,c)
    plt.title(dec.get(p[0],'UnKnown'))
    plt.imshow(img,cmap='gray')
    plt.axis('off')
    c+=1
  else:
    print(f"Fail to load image:{i}")
    plt.show()
```
